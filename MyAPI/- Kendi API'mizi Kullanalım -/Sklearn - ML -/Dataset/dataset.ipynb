```python
# ==========================================================
# ðŸ  Basit Regresyon Veri Seti (Ev FiyatÄ±)
# ==========================================================
import pandas as pd
import numpy as np

np.random.seed(42)

# Ã–zellikler: metrekare, oda sayÄ±sÄ±, bina yaÅŸÄ±
n = 200
metrekare = np.random.randint(60, 300, n)
oda_sayisi = np.random.randint(1, 6, n)
bina_yasi = np.random.randint(0, 30, n)

# GerÃ§ek fiyat = 2000 * metrekare + 50000 * oda_sayisi - 3000 * bina_yasi + gÃ¼rÃ¼ltÃ¼
fiyat = 2000*metrekare + 50000*oda_sayisi - 3000*bina_yasi + np.random.randint(-50000, 50000, n)

df = pd.DataFrame({
    "metrekare": metrekare,
    "oda_sayisi": oda_sayisi,
    "bina_yasi": bina_yasi,
    "fiyat": fiyat
})

df.to_csv("dataset_regresyon.csv", index=False)
print("âœ… Veri seti oluÅŸturuldu â†’ dataset_regresyon.csv")
print(df.head())
```

```python
# ==========================================================
# ðŸ’³ SÄ±nÄ±flandÄ±rma Veri Seti (Kredi OnayÄ±)
# ==========================================================
import pandas as pd
import numpy as np

np.random.seed(42)
n = 300

gelir = np.random.randint(3000, 15000, n)
yas = np.random.randint(18, 65, n)
borc_orani = np.random.uniform(0.1, 0.9, n)

# 1: OnaylandÄ±, 0: Reddedildi
onay = ((gelir > 8000) & (borc_orani < 0.4) & (yas > 22)).astype(int)

df = pd.DataFrame({
    "gelir": gelir,
    "yas": yas,
    "borc_orani": borc_orani,
    "onay": onay
})

df.to_csv("dataset_kredi.csv", index=False)
print("âœ… Veri seti oluÅŸturuldu â†’ dataset_kredi.csv")
print(df.head())
```
