```python
from hyso.models import EBO , CEMO , ELO
from hyso.augmentation import get_loaders
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import os
```

```python
data_dir = r"C:\Users\hdgn5\OneDrive\Masaüstü\MLOps\API - (Model Servişleştirme)\Html ve Css 'e Bağlama\Kendi CNN PyTorch Kütüphanemiz İle\split_data\test"  # senin ana klasör yolun
folders = [f for f in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, f))]

print("Toplam sınıf (klasör) sayısı:", len(folders))
print("Klasörler:", folders)
```

```python
# ---------------- LOADERS (hyso ile) ----------------
train_loader, val_loader = get_loaders(
    train_path=r"C:\Users\hdgn5\OneDrive\Masaüstü\MLOps\API - (Model Servişleştirme)\Html ve Css 'e Bağlama\Kendi CNN PyTorch Kütüphanemiz İle\split_data\train",   # ✅ train dataset
    val_path=r"C:\Users\hdgn5\OneDrive\Masaüstü\MLOps\API - (Model Servişleştirme)\Html ve Css 'e Bağlama\Kendi CNN PyTorch Kütüphanemiz İle\split_data\test",      # ✅ test dataset
    batch_size=64,
    img_size=(32,32),
    augment="advanced",
    channels=3
)
```

```python

# ---------------- MODEL ----------------
model = EBO(input_channels=3, num_classes=36)
model.callbacks(
    optimizer="Adam",
    scheduler="StepLR",
    scheduler_params={"step_size":5, "gamma":0.5},
    model_checkpoint=True,
    earlystop={"patience":5, "verbose":True}
)
```

```python
# ---------------- TRAINING ----------------
model.fit(
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=1
)
```
